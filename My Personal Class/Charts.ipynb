# 10 DIFFERENT TYPES OF CHARTS

```python
#LINECHART

import matplotlib.pyplot as plt

x = [10,20,30,40,50]
y = [15, 30,25, 5, 10]

plt.plot(x,y)
plt.title("Treasure's Line Chart")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
#BARCHART

import matplotlib.pyplot as plt

Categories = ['A', 'B', 'C', 'D', 'E']
Values = ['75', '40', '60', '50', '80']

plt.bar(Categories, Values)
plt.title("Treasure's Bar Chart")
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#PIE CHART 1

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.title("Pie Chart")
plt.show()
```

```python
#PIECHART 2

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ['Apples', 'Banana', 'Cherries', 'Dates']

plt.pie(y, labels = mylabels)
plt.show()
```

```python
#PIECHART 3

import matplotlib.pyplot as plt

labels = ['Money', 'Car', 'Travel']
Values = [30, 5, 60]

plt.pie(Values, labels=labels, autopct="%1.1f%%")
plt.title('Pie Chart')
#plt.legend()
plt.show()
```

```python
#PIECHART 4 
#Exlpoding the Pie Chart

import matplotlib.pyplot as plt

labels = ['Money', 'Car', 'Travel']
Values = [30, 5, 60]
myexplode = [0.1, 0.2, 0]

plt.pie(Values, labels=labels, autopct="%1.1f%%" , explode = myexplode)
plt.title('Pie Chart')
#plt.legend()
plt.show()
```

```python
#HISTOGRAM

import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(100)

plt.hist(data, bins=25, edgecolor='red')
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel('X-Axis')
plt.ylabel("Y-Axis")
plt.show()
```
